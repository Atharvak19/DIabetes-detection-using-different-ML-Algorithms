```python
from sklearn.preprocessing import StandardScaler
```

```python
SS = StandardScaler()
```

```python
SS_T = SS.transform(diab_new.drop('Outcome',axis=1))
```

```python
diab_new_ = pd.DataFrame(SS_T,columns=diab_new.columns[:-1])
```

```python
X_train, X_test, y_train, y_test = train_test_split(SS_T, diab_new['Outcome'], test_size=0.3, random_state=101)
```

```python
error = []

for i in range(1,50):
    KNN_i=KNeighborsClassifier(n_neighbors=i)
    KNN_i.fit(X_train,y_train)
    pred_i = KNN_i.predict(X_test)
    
    error.append(np.mean(y_test != pred_i))
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,50),error, ls='--',color= 'green',lw=2, marker = '*', markerfacecolor = 'purple',markersize=10)
plt.xlabel = ('range')
plt.ylabel = ('error rate')
plt.title = (' range vs error rate')
```

```python
KNN = KNeighborsClassifier(n_neighbors=9)
```
